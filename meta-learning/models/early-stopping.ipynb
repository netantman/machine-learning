As a rule of thumb, when the ML algorithm underfits, it helps (or does not hurt) to check if the optimization has converged. On the other hand, if it tends to overfit, **early stopping will help avoid getting into narrow optimum**.

Early stopping is the tricks that are favored in ML algorithms that uses SGD, such as [deep learning](deep-learning-debugging-and-visualization.ipynb). The way is to stop at an epoch where the validation error has seen the minimum for a while, as exemplified by the following:



Note that in the graph above, the x axis is the epoch, so it is **neighter learning curve or validation curve**.
- Not to confused by the learning curves above where the x axis is the size of the training data set. 
- Nor is the above the validation curve, whose x axis is the varied values of some hyperparameter.

More generally, if one views epoch as merely another hyperparameter, **it can be seen as a form of regularization**, just as tree size regularizes trees, etc.
