# Kernel Density Estimation

## Model Specification

One approach to estimate density from data is to make use of the histogram:

\begin{align}
\hat{f}_X(x_0) = \frac{\sum_{i=1}^N1(x_i\in N(x_0))}{N\lambda},
\end{align}
where $N(x_0)$ is a neighborhood around $x_0$ with width $\lambda$. The estimate is model-free, but it is bumpy. An alternative is to make use of kernels, or the smooth **Parzen estimate**

\begin{align}
\hat{f}_X(x_0) = \frac{1}{N\lambda}K_{\lambda}(x_0, x_i).
\end{align}
This is smoother, but the drawback is that the result is not sensitive to the choice of kernel and parameter $\lambda$.



**Available Kernels**

- Gaussian kernel (most popular): $K(x; h) \propto \exp(-\frac{x^2}{2h^2})$.
- Tophat kernel: $K(x; h) \propto 1$ when $|x|<h$.
- Epanechnikov kernel: $K(x; h) \propto 1-\frac{x^2}{h^2}$.
- Exponential kernel: $K(x; h) \propto\exp(-x/h)$.
- Linear kernel: $K(x; h) \propto 1-x/h$ when $x<h$.
- Cosine kernel: $K(x; h) \propto \cos\left(\frac{\pi x}{2h}\right)$ when $x<h$.

## Implementation Details and Practical Tricks

**`KernelDensity` in `sklearn`**

```python
from sklearn.neighbors.kde import KernelDensity
import numpy as np
X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
kde = KernelDensity(kernel='gaussian', bandwidth=1.0)
kde.fit(X)
kde.score_samples(X)
```

**Selected Parameters**

- **`bandwidth`**: 

    The bandwidth of the kernel.
    

- **`kernel`**:

    The kernel to use. Valid kernels are `[‘gaussian’|’tophat’|’epanechnikov’|’exponential’|’linear’|’cosine’]` Default is `‘gaussian’`.

**Selected Methods**

- `sample([n_samples, random_state])`: Generate random samples from the model.
- `score(X[, y])`: Compute the total log probability under the model.
- `score_samples(X)`: Evaluate the density model on the data.
